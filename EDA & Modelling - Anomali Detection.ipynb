# Modelling

## PCA 

```python
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import euclidean_distances

def get_euclidean_distance(x, centroids):
    return euclidean_distances([x.values], [centroids])[0][0]

def get_percentile(x, icr=10):
    df_pct = []
    if type(icr) == int:
        for i in range(icr, 100, icr):
            df_pct.append({'percentile':i, 'value':np.percentile(x, i)})
    else:
        icr = int(icr*10)
        for i in range(icr, 100*10, icr):
            df_pct.append({'percentile':i/10, 'value':np.percentile(x, i/10)})
    return pd.DataFrame(df_pct)
```

### model fitting 

```python
pca = PCA(2)
data = pca.fit_transform(df)
df['x'] = data[:, 0]
df['y'] = data[:, 1]
centroids = (sum(data[:, 0])/len(data), sum(data[:, 1])/len(data))
plt.scatter(x=data[:, 0], y=data[:, 1])
plt.scatter(centroids[0], centroids[1], marker="x", color='r', s=70,linewidth=3)
```

```python
print(centroids)
```

```python
df['euclidean_distance']= df[['x', 'y']].apply(lambda x : get_euclidean_distance(x, np.array(centroids)), axis=1)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 8))
sns.distplot(df['euclidean_distance'], hist=True, bins=5, ax=ax[0])
sns.boxplot(df['euclidean_distance'], ax=ax[1])
```

```python
get_percentile(df['euclidean_distance'], 0.5).tail(10)
```

### results 

```python
th =  0.14
df['flag_anomali'] = np.where(df['euclidean_distance']>=th, 1, 0)
```

```python
anomali_start_time = df[df['flag_anomali']==1].index.min()
print('anomali_start_time', anomali_start_time)
ax = df['euclidean_distance'].plot(figsize=(20, 8), title='euclidean_distance overtime')
ax.axhline(th, color='r', linewidth=1.5, linestyle ='--')
ax.axvline(anomali_start_time, color='k', linewidth=1.5, linestyle ='--')
ax.set_ylabel('euclidean_distance')
ax.set_xlabel('date')
```

```python
nb_epochs = 100
batch_size = 10
history = model_lstm.fit(X, X, epochs=nb_epochs, batch_size=batch_size,
                    validation_split=0.05).history
```

```python
fig, ax = plt.subplots(figsize=(14, 6), dpi=80)
ax.plot(history['loss'], 'b', label='Train', linewidth=2)
ax.plot(history['val_loss'], 'r', label='Validation', linewidth=2)
ax.set_title('Model loss', fontsize=16)
ax.set_ylabel('Loss (mae)')
ax.set_xlabel('Epoch')
ax.legend(loc='upper right')
plt.show()
```
